```python
!pip install kagglehub -q

import os
import kagglehub
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.manifold import TSNE

print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices('GPU'))
```

```python
path = kagglehub.dataset_download("msambare/fer2013")
print("Dataset path:", path)
```

```python
EMOTION_LABELS = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
train_dir = os.path.join(path, 'train')
test_dir  = os.path.join(path, 'test')

print("Training set distribution:")
for emo in EMOTION_LABELS:
    emo_path = os.path.join(train_dir, emo)
    if os.path.exists(emo_path):
        print(f"  {emo:10s}: {len(os.listdir(emo_path))} images")
```

```python
IMG_SIZE   = 96
LATENT_DIM = 512
BATCH_SIZE = 64
EPOCHS     = 60

print(f"Image Size : {IMG_SIZE}x{IMG_SIZE}")
print(f"Latent Dim : {LATENT_DIM}")
print(f"Max Epochs : {EPOCHS}")
```

```python

def collect_image_paths(base_dir, emotion_labels):
    paths, labels = [], []
    for idx, emo in enumerate(emotion_labels):
        emo_dir = os.path.join(base_dir, emo)
        if not os.path.exists(emo_dir):
            continue
        for fname in os.listdir(emo_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                paths.append(os.path.join(emo_dir, fname))
                labels.append(idx)
    return np.array(paths), np.array(labels)

train_paths, train_labels = collect_image_paths(train_dir, EMOTION_LABELS)
test_paths,  test_labels  = collect_image_paths(test_dir,  EMOTION_LABELS)

print(f"Train: {len(train_paths)} | Test: {len(test_paths)}")
```

```python
def load_image(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=1, expand_animations=False)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    return img

train_ds = tf.data.Dataset.from_tensor_slices(train_paths)
train_ds = train_ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(5000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print("Batch shape:", next(iter(train_ds)).shape)
```

```python
EMOTION_COLORS = {
    'angry':    '#E63946',
    'disgust':  '#8B4513',
    'fear':     '#6A0572',
    'happy':    '#F4A261',
    'neutral':  '#457B9D',
    'sad':      '#1D3557',
    'surprise': '#2A9D8F'
}

fig, axes = plt.subplots(3, 7, figsize=(20, 9))
fig.patch.set_facecolor('#F0F0F0')
fig.suptitle("FER2013 — Sample Faces per Emotion Category",
             fontsize=18, fontweight='bold', y=1.01)

for col, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == col]
    # Colored header bar
    axes[0, col].set_facecolor(EMOTION_COLORS[emo])
    axes[0, col].text(0.5, 0.5, emo.upper(),
                      ha='center', va='center',
                      fontsize=11, fontweight='bold', color='white',
                      transform=axes[0, col].transAxes)
    axes[0, col].axis('off')
    # Two sample images
    for row in range(1, 3):
        img = load_image(emo_paths[row]).numpy().squeeze()
        axes[row, col].imshow(img, cmap='gray', vmin=0, vmax=1)
        axes[row, col].axis('off')
        # Colored border
        for spine in axes[row, col].spines.values():
            spine.set_edgecolor(EMOTION_COLORS[emo])
            spine.set_linewidth(3)
            spine.set_visible(True)

plt.tight_layout()
plt.savefig('01_sample_images.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: 01_sample_images.png")
```

```python
encoder_inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 1), name='encoder_input')

x = layers.Conv2D(32,  3, strides=2, padding='same', use_bias=False)(encoder_inputs)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)

x = layers.Conv2D(64,  3, strides=2, padding='same', use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)

x = layers.Conv2D(128, 3, strides=2, padding='same', use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)

x = layers.Conv2D(256, 3, strides=2, padding='same', use_bias=False)(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)

x = layers.Flatten()(x)
x = layers.Dense(1024)(x)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Dropout(0.2)(x)   # light dropout prevents over-smoothing

z_mean    = layers.Dense(LATENT_DIM, name='z_mean',
                kernel_initializer='glorot_uniform')(x)
z_log_var = layers.Dense(LATENT_DIM, name='z_log_var',
                kernel_initializer='glorot_uniform')(x)

def sampling(args):
    z_mean, z_log_var = args
    z_log_var_clipped = tf.clip_by_value(z_log_var, -4, 4)  # wider clip
    epsilon = tf.random.normal(shape=tf.shape(z_mean))
    return z_mean + tf.exp(0.5 * z_log_var_clipped) * epsilon

z = layers.Lambda(sampling, name='z')([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name='encoder')
encoder.summary()
```

```python
latent_inputs = keras.Input(shape=(LATENT_DIM,), name='latent_input')

x = layers.Dense(6 * 6 * 256, kernel_initializer='glorot_uniform')(latent_inputs)
x = layers.BatchNormalization()(x)
x = layers.LeakyReLU(0.2)(x)
x = layers.Reshape((6, 6, 256))(x)

# Each block: upsample to 2x, then refine with a Conv2D
# This avoids Conv2DTranspose checkerboard artifacts entirely
def upsample_block(x, filters):
    x = layers.UpSampling2D(size=(2, 2), interpolation='bilinear')(x)
    x = layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    # Second conv to add representational capacity
    x = layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    return x

x = upsample_block(x, 256)  # 6x6 → 12x12
x = upsample_block(x, 128)  # 12x12 → 24x24
x = upsample_block(x, 64)   # 24x24 → 48x48
x = upsample_block(x, 32)   # 48x48 → 96x96

# Final projection to 1-channel image
decoder_outputs = layers.Conv2D(1, 3, padding='same', activation='sigmoid')(x)

decoder = keras.Model(latent_inputs, decoder_outputs, name='decoder')
decoder.summary()
```

```python
class EmotionVAE(keras.Model):
    def __init__(self, encoder, decoder, beta=0.001, ssim_weight=0.7, **kwargs):
        super().__init__(**kwargs)
        self.encoder     = encoder
        self.decoder     = decoder
        self.beta        = beta
        self.ssim_weight = ssim_weight   # 0.7 SSIM + 0.3 MSE
        self.total_loss_tracker = keras.metrics.Mean(name='total_loss')
        self.recon_loss_tracker = keras.metrics.Mean(name='recon_loss')
        self.kl_loss_tracker    = keras.metrics.Mean(name='kl_loss')

    @property
    def metrics(self):
        return [self.total_loss_tracker,
                self.recon_loss_tracker,
                self.kl_loss_tracker]

    def call(self, inputs, training=False):
        z_mean, z_log_var, z = self.encoder(inputs, training=training)
        return self.decoder(z, training=training)

    def _recon_loss(self, original, reconstruction):
        # SSIM loss: 1 - SSIM so that lower = better
        ssim_val  = tf.image.ssim(original, reconstruction,
                                  max_val=1.0, filter_size=7)
        ssim_loss = tf.reduce_mean(1.0 - ssim_val)

        # MSE loss (pixel-level detail)
        mse_loss  = tf.reduce_mean(tf.square(original - reconstruction))

        # Hybrid: SSIM captures structure, MSE captures fine detail
        return self.ssim_weight * ssim_loss + (1 - self.ssim_weight) * mse_loss

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction       = self.decoder(z, training=True)

            recon_loss = self._recon_loss(data, reconstruction)

            z_log_var_c = tf.clip_by_value(z_log_var, -4, 4)
            z_mean_c    = tf.clip_by_value(z_mean,    -10, 10)
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var_c - tf.square(z_mean_c) - tf.exp(z_log_var_c)
            )

            total_loss = recon_loss + self.beta * kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        grads = [tf.clip_by_norm(g, 1.0)
                 if g is not None else g for g in grads]
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {m.name: m.result() for m in self.metrics}
```

```python
# Cosine decay: starts high, decays smoothly — avoids getting stuck
total_steps = (len(train_paths) // BATCH_SIZE) * EPOCHS
lr_schedule = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-4,
    decay_steps=total_steps,
    alpha=1e-6           # minimum LR at end
)

optimizer = keras.optimizers.Adam(
    learning_rate=lr_schedule,
    epsilon=1e-7
)

early_stopping = keras.callbacks.EarlyStopping(
    monitor='total_loss', patience=8,   # more patience since LR decays
    restore_best_weights=True, verbose=1, mode='min'
)

vae = EmotionVAE(encoder, decoder, beta=0.001, ssim_weight=0.7)
vae.compile(optimizer=optimizer)

history = vae.fit(
    train_ds, epochs=EPOCHS,
    callbacks=[early_stopping],
    verbose=1
)

print(f"Stopped at epoch : {len(history.history['total_loss'])}")
print(f"Best total loss  : {min(history.history['total_loss']):.4f}")
```

```python

PROTOTYPE_N = 300
emotion_prototypes = {}

print("Computing emotion prototype vectors...")
for idx, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == idx]
    n         = min(PROTOTYPE_N, len(emo_paths))
    imgs      = np.array([load_image(p).numpy() for p in emo_paths[:n]])
    z_m, _, _ = encoder.predict(imgs, verbose=0)
    emotion_prototypes[emo] = np.mean(z_m, axis=0)
    print(f"  {emo:10s}: done ({n} images)")

print("\nAll prototypes ready!")
```

```python
N_PER_CLASS = 200
sample_paths, sample_labels_idx = [], []

for idx in range(len(EMOTION_LABELS)):
    emo_paths = test_paths[test_labels == idx]
    chosen    = emo_paths[:N_PER_CLASS]
    sample_paths.extend(chosen)
    sample_labels_idx.extend([idx] * len(chosen))

sample_imgs = np.array([load_image(p).numpy() for p in sample_paths])
z_means, _, _ = encoder.predict(sample_imgs, verbose=0)

print("Running t-SNE...")
tsne = TSNE(n_components=2, perplexity=40, random_state=42, n_iter=1000)
z_2d = tsne.fit_transform(z_means)
print("Done!")

fig, ax = plt.subplots(figsize=(14, 10))
ax.set_facecolor('#F0F4F8')
fig.patch.set_facecolor('#F0F4F8')

for idx, emo in enumerate(EMOTION_LABELS):
    mask = np.array(sample_labels_idx) == idx
    ax.scatter(
        z_2d[mask, 0], z_2d[mask, 1],
        c=EMOTION_COLORS[emo], label=emo.capitalize(),
        alpha=0.75, s=60,
        edgecolors='white', linewidths=0.5
    )
    # Add emotion label at cluster center
    cx = z_2d[mask, 0].mean()
    cy = z_2d[mask, 1].mean()
    ax.annotate(emo.upper(),
                xy=(cx, cy),
                fontsize=12, fontweight='bold',
                color=EMOTION_COLORS[emo],
                ha='center',
                bbox=dict(boxstyle='round,pad=0.3',
                         facecolor='white', alpha=0.7,
                         edgecolor=EMOTION_COLORS[emo]))

ax.legend(fontsize=12, title='Emotion', title_fontsize=13,
          framealpha=0.9, loc='upper right')
ax.set_title('t-SNE Visualization of Latent Space\n'
             'Each dot = one face | Colors = emotion category | '
             'Clusters show model learned emotion differences',
             fontsize=14, fontweight='bold')
ax.set_xlabel('t-SNE Dimension 1', fontsize=12)
ax.set_ylabel('t-SNE Dimension 2', fontsize=12)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('04_tsne.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: 04_tsne.png")
```

```python
# This is your key result — shows emotion transfer clearly
# Each row: one base face being transformed step by step

transfers = [
    ('happy',   'neutral',  'sad'),
    ('neutral', 'happy',    'surprise'),
    ('sad',     'neutral',  'happy'),
    ('angry',   'neutral',  'happy'),
]

alphas = [0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5]

fig, axes = plt.subplots(len(transfers), len(alphas),
                          figsize=(24, 4 * len(transfers)))
fig.patch.set_facecolor('#1a1a2e')
fig.suptitle('EMOTION TRANSFER — Latent Space Vector Arithmetic\n'
             'Each row shows gradual transformation from source emotion to target emotion',
             fontsize=16, fontweight='bold', color='white', y=1.01)

for row, (src, mid, tgt) in enumerate(transfers):
    src_path = train_paths[train_labels == EMOTION_LABELS.index(src)][10]
    img      = load_image(src_path).numpy()
    z_base, _, _ = encoder.predict(np.expand_dims(img, 0), verbose=0)
    direction    = emotion_prototypes[tgt] - emotion_prototypes[src]

    for col, a in enumerate(alphas):
        z_new = z_base + a * direction
        gen   = decoder.predict(z_new, verbose=0)[0].squeeze()

        ax = axes[row, col]
        ax.imshow(gen, cmap='gray', vmin=0, vmax=1)
        ax.axis('off')

        # Title for top row only
        if row == 0:
            ax.set_title(f'α = {a}', fontsize=11,
                        color='white', fontweight='bold')

        # Colored border based on progress
        border_color = EMOTION_COLORS[src] if a == 0.0 else \
                       EMOTION_COLORS[tgt] if a >= 1.0 else 'white'
        for spine in ax.spines.values():
            spine.set_edgecolor(border_color)
            spine.set_linewidth(3)
            spine.set_visible(True)

    # Row label on left
    axes[row, 0].set_ylabel(
        f'{src.upper()} → {tgt.upper()}',
        fontsize=12, fontweight='bold',
        color=EMOTION_COLORS[src],
        rotation=90, labelpad=10
    )

# Add alpha legend at bottom
fig.text(0.5, -0.02,
         'α=0.0 (original)  →  α=0.5 (halfway)  →  α=1.0 (full transfer)  →  α=1.5 (exaggerated)',
         ha='center', fontsize=12, color='white', style='italic')

plt.tight_layout()
plt.savefig('05_emotion_transfer.png', dpi=150,
            bbox_inches='tight', facecolor='#1a1a2e')
plt.show()
print("Saved: 05_emotion_transfer.png")
```
