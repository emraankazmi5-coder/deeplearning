```python
!pip install kagglehub -q

import os
import kagglehub
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.manifold import TSNE

print("TensorFlow version:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("msambare/fer2013")

print("Path to dataset files:", path)
```

```python
EMOTION_LABELS = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
train_dir = os.path.join(path, 'train')
test_dir  = os.path.join(path, 'test')

print("Training set distribution:")
for emo in EMOTION_LABELS:
    emo_path = os.path.join(train_dir, emo)
    if os.path.exists(emo_path):
        count = len(os.listdir(emo_path))
        print(f"  {emo:10s}: {count} images")
```

```python
IMG_SIZE   = 48
LATENT_DIM = 128
BATCH_SIZE = 64
EPOCHS     = 50

def collect_image_paths(base_dir, emotion_labels):
    paths, labels = [], []
    for idx, emo in enumerate(emotion_labels):
        emo_dir = os.path.join(base_dir, emo)
        if not os.path.exists(emo_dir):
            continue
        for fname in os.listdir(emo_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                paths.append(os.path.join(emo_dir, fname))
                labels.append(idx)
    return np.array(paths), np.array(labels)

train_paths, train_labels = collect_image_paths(train_dir, EMOTION_LABELS)
test_paths,  test_labels  = collect_image_paths(test_dir,  EMOTION_LABELS)

print(f"Total train images: {len(train_paths)}")
print(f"Total test  images: {len(test_paths)}")
```

```python
def load_image(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=1, expand_animations=False)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    return img

train_ds = tf.data.Dataset.from_tensor_slices(train_paths)
train_ds = train_ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(5000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print("Sample batch shape:", next(iter(train_ds)).shape)
```

```python
fig, axes = plt.subplots(2, 7, figsize=(18, 6))
fig.suptitle("FER2013 — Sample Images per Emotion", fontsize=16, fontweight='bold')

for col, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == col]
    for row in range(2):
        img = load_image(emo_paths[row]).numpy().squeeze()
        axes[row, col].imshow(img, cmap='gray')
        axes[row, col].axis('off')
        if row == 0:
            axes[row, col].set_title(emo.capitalize(), fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
encoder_inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 1), name='encoder_input')

x = layers.Conv2D(32,  3, activation='relu', strides=2, padding='same')(encoder_inputs)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(64,  3, activation='relu', strides=2, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Conv2D(128, 3, activation='relu', strides=2, padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Flatten()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.3)(x)

z_mean    = layers.Dense(LATENT_DIM, name='z_mean')(x)
z_log_var = layers.Dense(LATENT_DIM, name='z_log_var')(x)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.random.normal(shape=tf.shape(z_mean))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling, name='z')([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name='encoder')
encoder.summary()
```

```python
# After 3 stride-2 convolutions: 48 -> 24 -> 12 -> 6
latent_inputs = keras.Input(shape=(LATENT_DIM,), name='latent_input')

x = layers.Dense(6 * 6 * 128, activation='relu')(latent_inputs)
x = layers.Reshape((6, 6, 128))(x)
x = layers.Conv2DTranspose(128, 3, activation='relu', strides=2, padding='same')(x)  # -> 12
x = layers.BatchNormalization()(x)
x = layers.Conv2DTranspose(64,  3, activation='relu', strides=2, padding='same')(x)  # -> 24
x = layers.BatchNormalization()(x)
x = layers.Conv2DTranspose(32,  3, activation='relu', strides=2, padding='same')(x)  # -> 48
decoder_outputs = layers.Conv2DTranspose(1, 3, activation='sigmoid', padding='same')(x)

decoder = keras.Model(latent_inputs, decoder_outputs, name='decoder')
decoder.summary()
```

```python
class EmotionVAE(keras.Model):
    def __init__(self, encoder, decoder, beta=0.0005, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.beta    = beta
        self.total_loss_tracker = keras.metrics.Mean(name='loss')
        self.recon_loss_tracker = keras.metrics.Mean(name='recon_loss')
        self.kl_loss_tracker    = keras.metrics.Mean(name='kl_loss')

    @property
    def metrics(self):
        return [self.total_loss_tracker, self.recon_loss_tracker, self.kl_loss_tracker]

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        return self.decoder(z)

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction = self.decoder(z, training=True)

            recon_loss = tf.reduce_mean(
    tf.reduce_sum(
        tf.square(data - reconstruction),
        axis=(1, 2, 3)
    )
)
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
            )
            total_loss = recon_loss + self.beta * kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {m.name: m.result() for m in self.metrics}
```

```python
lr_schedule = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-3,
    decay_steps=EPOCHS * len(train_paths) // BATCH_SIZE
)

vae = EmotionVAE(encoder, decoder, beta=0.0005)
vae.compile(optimizer=keras.optimizers.Adam(lr_schedule))

history = vae.fit(train_ds, epochs=EPOCHS, verbose=1)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
fig.suptitle('Training Loss Curves', fontsize=14, fontweight='bold')

for ax, key, color in zip(axes,
    ['loss', 'recon_loss', 'kl_loss'],
    ['steelblue', 'darkorange', 'green']):
    ax.plot(history.history[key], color=color, linewidth=2)
    ax.set_title(key.replace('_', ' ').title())
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(3, 7, figsize=(20, 9))
fig.suptitle('Original (top) vs Reconstructed (middle) vs Difference (bottom)',
             fontsize=13, fontweight='bold')

for col, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == col]
    img   = load_image(emo_paths[0]).numpy()
    inp   = np.expand_dims(img, 0)
    recon = vae.predict(inp, verbose=0)[0]

    original      = img.squeeze()
    reconstructed = recon.squeeze()
    diff          = np.abs(original - reconstructed)

    axes[0, col].imshow(original,      cmap='gray', vmin=0, vmax=1)
    axes[1, col].imshow(reconstructed, cmap='gray', vmin=0, vmax=1)
    axes[2, col].imshow(diff,          cmap='hot',  vmin=0, vmax=0.5)

    for row in range(3):
        axes[row, col].axis('off')
    axes[0, col].set_title(emo.capitalize(), fontsize=11, fontweight='bold')

axes[0, 0].set_ylabel('Original',      fontsize=10)
axes[1, 0].set_ylabel('Reconstructed', fontsize=10)
axes[2, 0].set_ylabel('|Difference|',  fontsize=10)

plt.tight_layout()
plt.show()
```

```python
N_PER_CLASS = 200

sample_paths, sample_labels_idx = [], []
for idx in range(len(EMOTION_LABELS)):
    emo_paths = test_paths[test_labels == idx]
    chosen = emo_paths[:N_PER_CLASS]
    sample_paths.extend(chosen)
    sample_labels_idx.extend([idx] * len(chosen))

sample_imgs = np.array([load_image(p).numpy() for p in sample_paths])
z_means, _, _ = encoder.predict(sample_imgs, verbose=0)

print(f"Encoded {len(z_means)} images — running t-SNE...")
tsne = TSNE(n_components=2, perplexity=40, random_state=42, n_iter=1000)
z_2d = tsne.fit_transform(z_means)
print("t-SNE complete.")

EMOTION_COLORS = {
    'angry':    '#E63946',
    'disgust':  '#8B4513',
    'fear':     '#6A0572',
    'happy':    '#F4A261',
    'neutral':  '#457B9D',
    'sad':      '#1D3557',
    'surprise': '#2A9D8F'
}

fig, ax = plt.subplots(figsize=(12, 9))
ax.set_facecolor('#F8F9FA')
fig.patch.set_facecolor('#F8F9FA')

for idx, emo in enumerate(EMOTION_LABELS):
    mask = np.array(sample_labels_idx) == idx
    ax.scatter(
        z_2d[mask, 0], z_2d[mask, 1],
        c=EMOTION_COLORS[emo], label=emo.capitalize(),
        alpha=0.7, s=40, edgecolors='white', linewidths=0.4
    )

ax.legend(fontsize=12, frameon=True, loc='upper right',
          title='Emotion', title_fontsize=12)
ax.set_title('t-SNE of VAE Latent Space (Colored by Emotion)', fontsize=16, fontweight='bold')
ax.set_xlabel('t-SNE Dimension 1', fontsize=12)
ax.set_ylabel('t-SNE Dimension 2', fontsize=12)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
PROTOTYPE_N = 300

emotion_prototypes = {}

for idx, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == idx]
    chosen    = emo_paths[:PROTOTYPE_N]
    imgs      = np.array([load_image(p).numpy() for p in chosen])
    z_m, _, _ = encoder.predict(imgs, verbose=0)
    emotion_prototypes[emo] = np.mean(z_m, axis=0)
    print(f"  {emo:10s}: prototype computed from {len(chosen)} images")

print("\nAll emotion prototypes computed!")
```

```python
def emotion_arithmetic(base_img_path, source_emo, target_emo, alpha=1.0):
    img = load_image(base_img_path).numpy()
    z_base, _, _ = encoder.predict(np.expand_dims(img, 0), verbose=0)
    direction = emotion_prototypes[target_emo] - emotion_prototypes[source_emo]
    z_new = z_base + alpha * direction
    gen   = decoder.predict(z_new, verbose=0)[0]
    return img.squeeze(), gen.squeeze()

experiments = [
    ('neutral', 'happy',    'Neutral → Happy'),
    ('happy',   'sad',      'Happy   → Sad'),
    ('neutral', 'surprise', 'Neutral → Surprise'),
]

alphas = [0.0, 0.5, 1.0, 1.5, 2.0]
fig, axes = plt.subplots(len(experiments), 5, figsize=(18, 3.5 * len(experiments)))
fig.suptitle('Emotion Vector Arithmetic in Latent Space', fontsize=16, fontweight='bold', y=1.02)

for row, (src_emo, tgt_emo, title) in enumerate(experiments):
    src_path = train_paths[train_labels == EMOTION_LABELS.index(src_emo)][42]

    for col, a in enumerate(alphas):
        orig, gen = emotion_arithmetic(src_path, src_emo, tgt_emo, alpha=a)
        axes[row, col].imshow(gen if a > 0 else orig, cmap='gray', vmin=0, vmax=1)
        axes[row, col].axis('off')
        label = f'α = {a}'
        if a == 0.0: label += '\n(original)'
        if a == 1.0: label += '\n(full transfer)'
        axes[row, col].set_title(label, fontsize=9)

    axes[row, 0].set_ylabel(title, fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
n_emo = len(EMOTION_LABELS)
fig, axes = plt.subplots(n_emo, n_emo, figsize=(16, 16))
fig.suptitle('Emotion Interpolation Grid (Row × Column = Blended Expression)',
             fontsize=15, fontweight='bold')

for i, emo_row in enumerate(EMOTION_LABELS):
    for j, emo_col in enumerate(EMOTION_LABELS):
        z_blend = 0.5 * emotion_prototypes[emo_row] + 0.5 * emotion_prototypes[emo_col]
        gen = decoder.predict(z_blend[np.newaxis, :], verbose=0)[0].squeeze()

        axes[i, j].imshow(gen, cmap='gray', vmin=0, vmax=1)
        axes[i, j].axis('off')

        if i == j:
            for spine in axes[i, j].spines.values():
                spine.set_edgecolor(EMOTION_COLORS[emo_row])
                spine.set_linewidth(3)
                spine.set_visible(True)

    axes[i, 0].set_ylabel(emo_row.capitalize(), fontsize=10, fontweight='bold')

for j, emo_col in enumerate(EMOTION_LABELS):
    axes[0, j].set_title(emo_col.capitalize(), fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
def plot_emotion_walk(source_emo, target_emo, n_steps=9):
    z_src = emotion_prototypes[source_emo]
    z_tgt = emotion_prototypes[target_emo]

    fig, axes = plt.subplots(1, n_steps, figsize=(22, 3))
    fig.suptitle(f'Latent Walk: {source_emo.capitalize()} → {target_emo.capitalize()}',
                 fontsize=14, fontweight='bold')

    for i, t in enumerate(np.linspace(0, 1, n_steps)):
        z_interp = (1 - t) * z_src + t * z_tgt
        gen = decoder.predict(z_interp[np.newaxis, :], verbose=0)[0].squeeze()
        axes[i].imshow(gen, cmap='gray', vmin=0, vmax=1)
        axes[i].axis('off')
        axes[i].set_title(f't = {t:.2f}', fontsize=9)

    for s in axes[0].spines.values():
        s.set_visible(True); s.set_edgecolor(EMOTION_COLORS[source_emo]); s.set_linewidth(3)
    for s in axes[-1].spines.values():
        s.set_visible(True); s.set_edgecolor(EMOTION_COLORS[target_emo]); s.set_linewidth(3)

    plt.tight_layout()
    plt.show()

plot_emotion_walk('angry',   'happy')
plot_emotion_walk('sad',     'surprise')
plot_emotion_walk('disgust', 'neutral')
```

```python
def emotion_analogy(emo_a, emo_b, emo_c):
    z_result = (emotion_prototypes[emo_a]
                - emotion_prototypes[emo_b]
                + emotion_prototypes[emo_c])
    return decoder.predict(z_result[np.newaxis, :], verbose=0)[0].squeeze()

analogies = [
    ('happy',    'neutral',  'sad',      'happy − neutral + sad = ?'),
    ('surprise', 'neutral',  'angry',    'surprise − neutral + angry = ?'),
    ('fear',     'neutral',  'happy',    'fear − neutral + happy = ?'),
]

fig, axes = plt.subplots(len(analogies), 5, figsize=(16, 4 * len(analogies)))
fig.suptitle('Emotion Analogies: A − B + C = ? (Latent Vector Arithmetic)',
             fontsize=15, fontweight='bold', y=1.02)

for row, (a, b, c, desc) in enumerate(analogies):
    for col_idx, emo, label in [(0, a, f'A: {a}'), (1, b, f'B: {b}'), (2, c, f'C: {c}')]:
        img = decoder.predict(emotion_prototypes[emo][np.newaxis, :], verbose=0)[0].squeeze()
        axes[row, col_idx].imshow(img, cmap='gray', vmin=0, vmax=1)
        axes[row, col_idx].set_title(label.capitalize(), fontsize=10,
                                      fontweight='bold', color=EMOTION_COLORS[emo])
        axes[row, col_idx].axis('off')

    axes[row, 3].text(0.5, 0.5, desc, ha='center', va='center',
                      fontsize=10, fontweight='bold', color='gray',
                      transform=axes[row, 3].transAxes, wrap=True)
    axes[row, 3].axis('off')

    result_img = emotion_analogy(a, b, c)
    axes[row, 4].imshow(result_img, cmap='gray', vmin=0, vmax=1)
    axes[row, 4].set_title('Result (?)', fontsize=10, fontweight='bold')
    axes[row, 4].axis('off')

plt.tight_layout()
plt.show()
```
