```python
# ── 1. Dataset ───────────────────────────────────────────────────────────────
import kagglehub
path = kagglehub.dataset_download("msambare/fer2013")
print("Dataset path:", path)

EMOTION_LABELS = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
EMOTION_COLORS = {
    'angry':    '#E63946',
    'disgust':  '#8B4513',
    'fear':     '#6A0572',
    'happy':    '#F4A261',
    'neutral':  '#457B9D',
    'sad':      '#1D3557',
    'surprise': '#2A9D8F',
}

train_dir = os.path.join(path, 'train')
test_dir  = os.path.join(path, 'test')

print("\nTraining set distribution:")
for emo in EMOTION_LABELS:
    emo_path = os.path.join(train_dir, emo)
    if os.path.exists(emo_path):
        count = len(os.listdir(emo_path))
        print(f"  {emo:10s}: {count:5d} images")
```

```python
# ── 2. Hyperparameters ───────────────────────────────────────────────────────
IMG_SIZE      = 48
LATENT_DIM    = 256       # more expressive latent space
BATCH_SIZE    = 64
EPOCHS        = 60
BETA_MAX      = 0.001     # max KL weight (warm-up target)
WARMUP_EPOCHS = 15        # ramp KL weight from 0 → BETA_MAX over first N epochs
PROTOTYPE_N   = 400       # images per emotion for prototype computation
```

```python

# ── 3. Image loading helpers ─────────────────────────────────────────────────
def collect_image_paths(base_dir, emotion_labels):
    paths, labels = [], []
    for idx, emo in enumerate(emotion_labels):
        emo_dir = os.path.join(base_dir, emo)
        if not os.path.exists(emo_dir):
            continue
        for fname in os.listdir(emo_dir):
            if fname.lower().endswith(('.png', '.jpg', '.jpeg')):
                paths.append(os.path.join(emo_dir, fname))
                labels.append(idx)
    return np.array(paths), np.array(labels)


train_paths, train_labels = collect_image_paths(train_dir, EMOTION_LABELS)
test_paths,  test_labels  = collect_image_paths(test_dir,  EMOTION_LABELS)

print(f"\nTotal train images: {len(train_paths)}")
print(f"Total test  images: {len(test_paths)}")


def load_image(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=1, expand_animations=False)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0
    return img


def augment_image(img):
    """Light augmentation — flip + small brightness jitter."""
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, max_delta=0.08)
    img = tf.clip_by_value(img, 0.0, 1.0)
    return img


# Training dataset WITH augmentation
train_ds = (
    tf.data.Dataset.from_tensor_slices(train_paths)
    .map(load_image,     num_parallel_calls=tf.data.AUTOTUNE)
    .map(augment_image,  num_parallel_calls=tf.data.AUTOTUNE)
    .shuffle(8000)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

print("Sample batch shape:", next(iter(train_ds)).shape)
```

```python
# ── 5. Model building ────────────────────────────────────────────────────────

# 5a. Reparameterisation as a proper Keras layer
class Sampling(layers.Layer):
    """Reparameterisation trick: z = mu + sigma * eps."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim   = tf.shape(z_mean)[1]
        eps   = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * eps


# 5b. Residual block helper
def residual_block(x, filters, name_prefix):
    """Conv → BN → ReLU → Conv → BN + skip."""
    shortcut = x
    x = layers.Conv2D(filters, 3, padding='same', name=f'{name_prefix}_c1')(x)
    x = layers.BatchNormalization(name=f'{name_prefix}_bn1')(x)
    x = layers.ReLU()(x)
    x = layers.Conv2D(filters, 3, padding='same', name=f'{name_prefix}_c2')(x)
    x = layers.BatchNormalization(name=f'{name_prefix}_bn2')(x)
    if shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, 1, padding='same',
                                 name=f'{name_prefix}_proj')(shortcut)
    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)
    return x


# 5c. Encoder
def build_encoder(img_size, latent_dim):
    inp = keras.Input(shape=(img_size, img_size, 1), name='encoder_input')

    x = layers.Conv2D(32, 3, strides=2, padding='same', activation='relu')(inp)   # 24×24
    x = layers.BatchNormalization()(x)
    x = residual_block(x, 32,  'enc_res1')

    x = layers.Conv2D(64, 3, strides=2, padding='same', activation='relu')(x)     # 12×12
    x = layers.BatchNormalization()(x)
    x = residual_block(x, 64,  'enc_res2')

    x = layers.Conv2D(128, 3, strides=2, padding='same', activation='relu')(x)    # 6×6
    x = layers.BatchNormalization()(x)
    x = residual_block(x, 128, 'enc_res3')

    x = layers.Conv2D(256, 3, strides=2, padding='same', activation='relu')(x)    # 3×3
    x = layers.BatchNormalization()(x)

    x = layers.Flatten()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.25)(x)

    z_mean    = layers.Dense(latent_dim, name='z_mean')(x)
    z_log_var = layers.Dense(latent_dim, name='z_log_var',
                             kernel_initializer='zeros')(x)
    z         = Sampling(name='z')([z_mean, z_log_var])

    return keras.Model(inp, [z_mean, z_log_var, z], name='encoder')


# 5d. Decoder
def build_decoder(latent_dim, img_size):
    inp = keras.Input(shape=(latent_dim,), name='latent_input')

    x = layers.Dense(3 * 3 * 256, activation='relu')(inp)
    x = layers.Reshape((3, 3, 256))(x)

    x = layers.Conv2DTranspose(256, 3, strides=2, padding='same', activation='relu')(x)  # 6×6
    x = layers.BatchNormalization()(x)

    x = layers.Conv2DTranspose(128, 3, strides=2, padding='same', activation='relu')(x)  # 12×12
    x = layers.BatchNormalization()(x)
    x = residual_block(x, 128, 'dec_res1')

    x = layers.Conv2DTranspose(64, 3, strides=2, padding='same', activation='relu')(x)   # 24×24
    x = layers.BatchNormalization()(x)
    x = residual_block(x, 64,  'dec_res2')

    x = layers.Conv2DTranspose(32, 3, strides=2, padding='same', activation='relu')(x)   # 48×48
    x = layers.BatchNormalization()(x)

    out = layers.Conv2DTranspose(1, 3, padding='same', activation='sigmoid')(x)

    return keras.Model(inp, out, name='decoder')


encoder = build_encoder(IMG_SIZE, LATENT_DIM)
decoder = build_decoder(LATENT_DIM, IMG_SIZE)
encoder.summary()
decoder.summary()
```

```python
# ── 6. Hybrid reconstruction loss: MSE + SSIM ───────────────────────────────
def ssim_loss(y_true, y_pred):
    """1 − SSIM (higher SSIM = lower loss)."""
    return 1.0 - tf.reduce_mean(
        tf.image.ssim(y_true, y_pred, max_val=1.0)
    )


def reconstruction_loss(y_true, y_pred, alpha=0.85):
    """alpha × MSE + (1-alpha) × SSIM_loss."""
    mse  = tf.reduce_mean(tf.reduce_sum(tf.square(y_true - y_pred), axis=(1, 2, 3)))
    ssim = ssim_loss(y_true, y_pred) * (IMG_SIZE * IMG_SIZE)
    return alpha * mse + (1.0 - alpha) * ssim
```

```python
# ── 7. VAE with KL warm-up ───────────────────────────────────────────────────
class EmotionVAE(keras.Model):
    def __init__(self, encoder, decoder, beta_max=0.001, warmup_epochs=15,
                 steps_per_epoch=1, **kwargs):
        super().__init__(**kwargs)
        self.encoder      = encoder
        self.decoder      = decoder
        self.beta_max     = beta_max
        self.warmup_steps = warmup_epochs * steps_per_epoch

        self.beta  = tf.Variable(0.0, trainable=False, dtype=tf.float32, name='beta')
        self._step = tf.Variable(0,   trainable=False, dtype=tf.int64,   name='step')

        self.total_loss_tracker = keras.metrics.Mean(name='loss')
        self.recon_loss_tracker = keras.metrics.Mean(name='recon_loss')
        self.kl_loss_tracker    = keras.metrics.Mean(name='kl_loss')
        self.beta_tracker       = keras.metrics.Mean(name='beta')

    @property
    def metrics(self):
        return [self.total_loss_tracker, self.recon_loss_tracker,
                self.kl_loss_tracker, self.beta_tracker]

    def call(self, inputs, training=False):
        z_mean, z_log_var, z = self.encoder(inputs, training=training)
        return self.decoder(z, training=training)

    def train_step(self, data):
        self._step.assign_add(1)
        ratio = tf.cast(self._step, tf.float32) / float(self.warmup_steps)
        self.beta.assign(tf.minimum(ratio * self.beta_max, self.beta_max))

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction       = self.decoder(z, training=True)

            recon = reconstruction_loss(data, reconstruction)
            kl    = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var),
                              axis=1)
            )
            total = recon + self.beta * kl

        grads = tape.gradient(total, self.trainable_weights)
        grads, _ = tf.clip_by_global_norm(grads, 5.0)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total)
        self.recon_loss_tracker.update_state(recon)
        self.kl_loss_tracker.update_state(kl)
        self.beta_tracker.update_state(self.beta)
        return {m.name: m.result() for m in self.metrics}


steps_per_epoch = len(train_paths) // BATCH_SIZE

lr_schedule = keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=1e-3,
    decay_steps=EPOCHS * steps_per_epoch,
    alpha=1e-5,
)

vae = EmotionVAE(encoder, decoder,
                 beta_max=BETA_MAX,
                 warmup_epochs=WARMUP_EPOCHS,
                 steps_per_epoch=steps_per_epoch)

vae.compile(optimizer=keras.optimizers.Adam(lr_schedule, epsilon=1e-7))

history = vae.fit(train_ds, epochs=EPOCHS, verbose=1)
```

```python
# ── 10. Latent-space analysis: t-SNE + PCA ───────────────────────────────────
N_PER_CLASS = 200

sample_paths, sample_labels_idx = [], []
for idx in range(len(EMOTION_LABELS)):
    emo_paths = test_paths[test_labels == idx]
    chosen    = emo_paths[:N_PER_CLASS]
    sample_paths.extend(chosen)
    sample_labels_idx.extend([idx] * len(chosen))

sample_imgs       = np.array([load_image(p).numpy() for p in sample_paths])
z_means, _, _     = encoder.predict(sample_imgs, verbose=0)
sample_labels_arr = np.array(sample_labels_idx)

print(f"Encoded {len(z_means)} images")

print("Running t-SNE...")
tsne     = TSNE(n_components=2, perplexity=40, random_state=42,
                max_iter=1000, learning_rate='auto', init='pca')
z_2d_tsne = tsne.fit_transform(z_means)
print("t-SNE complete.")

pca      = PCA(n_components=2, random_state=42)
z_2d_pca = pca.fit_transform(z_means)
print(f"PCA explained variance: {pca.explained_variance_ratio_.sum() * 100:.1f}%")

fig, axes = plt.subplots(1, 2, figsize=(20, 8))
titles = [
    't-SNE of VAE Latent Space',
    f'PCA of VAE Latent Space\n(exp. var = {pca.explained_variance_ratio_.sum() * 100:.1f}%)',
]

for ax, z2d, title in zip(axes, [z_2d_tsne, z_2d_pca], titles):
    ax.set_facecolor('#F8F9FA')
    for idx, emo in enumerate(EMOTION_LABELS):
        mask = sample_labels_arr == idx
        ax.scatter(z2d[mask, 0], z2d[mask, 1],
                   c=EMOTION_COLORS[emo], label=emo.capitalize(),
                   alpha=0.7, s=40, edgecolors='white', linewidths=0.3)
    ax.legend(fontsize=10, title='Emotion', title_fontsize=10)
    ax.set_title(title, fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('latent_space.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 11. Emotion prototypes ────────────────────────────────────────────────────
emotion_prototypes = {}

for idx, emo in enumerate(EMOTION_LABELS):
    emo_paths = train_paths[train_labels == idx]
    chosen    = emo_paths[:PROTOTYPE_N]
    imgs      = np.array([load_image(p).numpy() for p in chosen])
    z_m, _, _ = encoder.predict(imgs, verbose=0)
    emotion_prototypes[emo] = np.mean(z_m, axis=0)
    print(f"  {emo:10s}: prototype from {len(chosen)} images")

print("\nAll emotion prototypes computed!")

# Prototype cosine-similarity heatmap
proto_matrix = np.stack([emotion_prototypes[e] for e in EMOTION_LABELS])
cos_sim      = 1 - pairwise_distances(proto_matrix, metric='cosine')

fig, ax = plt.subplots(figsize=(8, 7))
im = ax.imshow(cos_sim, cmap='RdYlGn', vmin=-0.5, vmax=1.0)
ax.set_xticks(range(len(EMOTION_LABELS)))
ax.set_yticks(range(len(EMOTION_LABELS)))
ax.set_xticklabels([e.capitalize() for e in EMOTION_LABELS], rotation=45, ha='right')
ax.set_yticklabels([e.capitalize() for e in EMOTION_LABELS])
for i in range(len(EMOTION_LABELS)):
    for j in range(len(EMOTION_LABELS)):
        ax.text(j, i, f'{cos_sim[i, j]:.2f}', ha='center', va='center',
                fontsize=9, color='black')
plt.colorbar(im, ax=ax, label='Cosine Similarity')
ax.set_title('Prototype Cosine-Similarity Heatmap\n(closer to 1 = more similar emotions)',
             fontweight='bold')
plt.tight_layout()
plt.savefig('prototype_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python

# ── 12. Emotion vector arithmetic ────────────────────────────────────────────
def emotion_arithmetic(base_img_path, source_emo, target_emo, alpha=1.0):
    img    = load_image(base_img_path).numpy()
    z_base, _, _ = encoder.predict(np.expand_dims(img, 0), verbose=0)
    direction = emotion_prototypes[target_emo] - emotion_prototypes[source_emo]
    z_new  = z_base + alpha * direction
    gen    = decoder.predict(z_new, verbose=0)[0]
    return img.squeeze(), gen.squeeze()


experiments = [
    ('neutral', 'happy',    'Neutral → Happy'),
    ('happy',   'sad',      'Happy → Sad'),
    ('neutral', 'surprise', 'Neutral → Surprise'),
    ('sad',     'angry',    'Sad → Angry'),
]

alphas = [0.0, 0.5, 1.0, 1.5, 2.0]
fig, axes = plt.subplots(len(experiments), 5,
                         figsize=(18, 3.5 * len(experiments)))
fig.suptitle('Emotion Vector Arithmetic in Latent Space',
             fontsize=16, fontweight='bold', y=1.01)

for row, (src_emo, tgt_emo, title) in enumerate(experiments):
    src_path = train_paths[train_labels == EMOTION_LABELS.index(src_emo)][42]
    for col, a in enumerate(alphas):
        orig, gen = emotion_arithmetic(src_path, src_emo, tgt_emo, alpha=a)
        axes[row, col].imshow(gen if a > 0 else orig, cmap='gray', vmin=0, vmax=1)
        axes[row, col].axis('off')
        label = f'α = {a}'
        if a == 0.0: label += '\n(original)'
        if a == 1.0: label += '\n(full)'
        axes[row, col].set_title(label, fontsize=9)
    axes[row, 0].set_ylabel(title, fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig('vector_arithmetic.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 13. Latent walk ───────────────────────────────────────────────────────────
def plot_emotion_walk(source_emo, target_emo, n_steps=9):
    z_src = emotion_prototypes[source_emo]
    z_tgt = emotion_prototypes[target_emo]

    fig, axes = plt.subplots(1, n_steps, figsize=(22, 3))
    fig.suptitle(f'Latent Walk: {source_emo.capitalize()} → {target_emo.capitalize()}',
                 fontsize=14, fontweight='bold')

    for i, t in enumerate(np.linspace(0, 1, n_steps)):
        z_interp = (1 - t) * z_src + t * z_tgt
        gen = decoder.predict(z_interp[np.newaxis, :], verbose=0)[0].squeeze()
        axes[i].imshow(gen, cmap='gray', vmin=0, vmax=1)
        axes[i].axis('off')
        axes[i].set_title(f't={t:.2f}', fontsize=9)

    for s in axes[0].spines.values():
        s.set_visible(True)
        s.set_edgecolor(EMOTION_COLORS[source_emo])
        s.set_linewidth(3)
    for s in axes[-1].spines.values():
        s.set_visible(True)
        s.set_edgecolor(EMOTION_COLORS[target_emo])
        s.set_linewidth(3)

    plt.tight_layout()
    plt.savefig(f'walk_{source_emo}_{target_emo}.png', dpi=150, bbox_inches='tight')
    plt.show()


plot_emotion_walk('angry',   'happy')
plot_emotion_walk('sad',     'surprise')
plot_emotion_walk('disgust', 'neutral')
plot_emotion_walk('fear',    'happy')
```

```python
# ── 14. Emotion analogies ─────────────────────────────────────────────────────
def emotion_analogy(emo_a, emo_b, emo_c):
    z_result = (emotion_prototypes[emo_a]
                - emotion_prototypes[emo_b]
                + emotion_prototypes[emo_c])
    return decoder.predict(z_result[np.newaxis, :], verbose=0)[0].squeeze()


analogies = [
    ('happy',    'neutral',  'sad',    'happy − neutral + sad'),
    ('surprise', 'neutral',  'angry',  'surprise − neutral + angry'),
    ('fear',     'neutral',  'happy',  'fear − neutral + happy'),
    ('sad',      'neutral',  'happy',  'sad − neutral + happy'),
]

fig, axes = plt.subplots(len(analogies), 5,
                         figsize=(16, 4 * len(analogies)))
fig.suptitle('Emotion Analogies  A − B + C = ?',
             fontsize=15, fontweight='bold', y=1.01)

for row, (a, b, c, desc) in enumerate(analogies):
    for col_idx, emo in enumerate([a, b, c]):
        img = decoder.predict(emotion_prototypes[emo][np.newaxis, :],
                              verbose=0)[0].squeeze()
        label = ['A', 'B', 'C'][col_idx]
        axes[row, col_idx].imshow(img, cmap='gray', vmin=0, vmax=1)
        axes[row, col_idx].set_title(f'{label}: {emo}'.capitalize(), fontsize=10,
                                     fontweight='bold',
                                     color=EMOTION_COLORS[emo])
        axes[row, col_idx].axis('off')

    axes[row, 3].text(0.5, 0.5, desc, ha='center', va='center',
                      fontsize=9, fontweight='bold', color='gray',
                      transform=axes[row, 3].transAxes, wrap=True)
    axes[row, 3].axis('off')

    result_img = emotion_analogy(a, b, c)
    axes[row, 4].imshow(result_img, cmap='gray', vmin=0, vmax=1)
    axes[row, 4].set_title('Result (?)', fontsize=10, fontweight='bold')
    axes[row, 4].axis('off')

plt.tight_layout()
plt.savefig('analogies.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 15. Average Reconstruction Error by Emotion ──────────────────────────────
print("=" * 65)
print("  15. Average Reconstruction Error by Emotion")
print("=" * 65)

N_EVAL = 150   # images per class

avg_mse_per_emo  = {}
avg_ssim_per_emo = {}

for idx, emo in enumerate(EMOTION_LABELS):
    emo_paths = test_paths[test_labels == idx][:N_EVAL]
    imgs      = np.array([load_image(p).numpy() for p in emo_paths])
    recons    = vae.predict(imgs, verbose=0)

    mse_vals  = np.mean((imgs - recons) ** 2, axis=(1, 2, 3))
    ssim_vals = np.array([
        float(tf.image.ssim(
            tf.constant(imgs[i]),
            tf.constant(recons[i]),
            max_val=1.0
        ).numpy())
        for i in range(len(imgs))
    ])

    avg_mse_per_emo[emo]  = (np.mean(mse_vals),  np.std(mse_vals))
    avg_ssim_per_emo[emo] = (np.mean(ssim_vals), np.std(ssim_vals))
    print(f"  {emo:10s}  MSE={avg_mse_per_emo[emo][0]:.5f} ± {avg_mse_per_emo[emo][1]:.5f}"
          f"   SSIM={avg_ssim_per_emo[emo][0]:.4f} ± {avg_ssim_per_emo[emo][1]:.4f}")

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle("Average Reconstruction Error by Emotion (Test Set)",
             fontsize=14, fontweight='bold')

for ax, metric, label, ylab in zip(
        axes,
        [avg_mse_per_emo, avg_ssim_per_emo],
        ['MSE', 'SSIM'],
        ['Mean Squared Error', 'Structural Similarity (SSIM)']):

    emos   = EMOTION_LABELS
    means  = [metric[e][0] for e in emos]
    stds   = [metric[e][1] for e in emos]
    colors = [EMOTION_COLORS[e] for e in emos]

    bars = ax.bar(emos, means, color=colors, yerr=stds,
                  capsize=5, edgecolor='white', linewidth=0.8,
                  error_kw={'ecolor': '#333', 'elinewidth': 1.2})
    ax.set_title(f'Per-Emotion {label}', fontsize=12, fontweight='bold')
    ax.set_ylabel(ylab)
    ax.set_xticklabels([e.capitalize() for e in emos], rotation=30, ha='right')
    ax.grid(axis='y', alpha=0.3)

    for bar, val, std in zip(bars, means, stds):
        ax.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + std + (max(means) * 0.01),
                f'{val:.4f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('recon_error_by_emotion.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 16. Latent-Space Confusion Matrix (k-NN) ─────────────────────────────────
print("=" * 65)
print("  16. Latent-Space Confusion Matrix (k-NN, k=5)")
print("=" * 65)

N_TRAIN_KNN = 300
N_TEST_KNN  = 150


def encode_split(paths_arr, labels_arr, n_per_class):
    X, y = [], []
    for idx in range(len(EMOTION_LABELS)):
        chosen = paths_arr[labels_arr == idx][:n_per_class]
        imgs   = np.array([load_image(p).numpy() for p in chosen])
        z_m, _, _ = encoder.predict(imgs, verbose=0)
        X.append(z_m)
        y.extend([idx] * len(chosen))
    return np.vstack(X), np.array(y)


print("  Encoding training set for k-NN …")
X_tr, y_tr = encode_split(train_paths, train_labels, N_TRAIN_KNN)

print("  Encoding test set for k-NN …")
X_te, y_te = encode_split(test_paths, test_labels, N_TEST_KNN)

knn    = KNeighborsClassifier(n_neighbors=5, metric='cosine', n_jobs=-1)
knn.fit(X_tr, y_tr)
y_pred = knn.predict(X_te)

cm         = confusion_matrix(y_te, y_pred)
cm_norm    = cm.astype(float) / cm.sum(axis=1, keepdims=True)
overall_acc = np.trace(cm) / cm.sum()

print(f"\n  k-NN Latent-Space Accuracy: {overall_acc * 100:.2f}%")
for idx, emo in enumerate(EMOTION_LABELS):
    print(f"    {emo:10s}: {cm_norm[idx, idx] * 100:.1f}%")

fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle(f"Latent-Space Confusion Matrix  (k-NN, k=5)\n"
             f"Overall Accuracy: {overall_acc * 100:.1f}%",
             fontsize=14, fontweight='bold')

labels_cap = [e.capitalize() for e in EMOTION_LABELS]

for ax, data, title, fmt in zip(
        axes,
        [cm, cm_norm],
        ['Raw Counts', 'Normalised (row %)'],
        ['d', '.2f']):

    cmap_cm = LinearSegmentedColormap.from_list('cm_cmap', ['#F8F9FA', '#1D3557'], N=256)
    im = ax.imshow(data, cmap=cmap_cm, aspect='auto',
                   vmin=0, vmax=(cm.max() if fmt == 'd' else 1.0))

    ax.set_xticks(range(len(EMOTION_LABELS)))
    ax.set_yticks(range(len(EMOTION_LABELS)))
    ax.set_xticklabels(labels_cap, rotation=45, ha='right', fontsize=10)
    ax.set_yticklabels(labels_cap, fontsize=10)
    ax.set_xlabel('Predicted', fontsize=11)
    ax.set_ylabel('True', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')

    thresh = data.max() / 2.0
    for i, j in itertools.product(range(data.shape[0]), range(data.shape[1])):
        val = data[i, j]
        txt = f'{val:{fmt}}' if fmt == 'd' else f'{val:.2f}'
        ax.text(j, i, txt, ha='center', va='center', fontsize=8,
                color='white' if val > thresh else 'black')
    plt.colorbar(im, ax=ax, shrink=0.85)

plt.tight_layout()
plt.savefig('latent_confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 17. Per-Emotion Latent Distribution (violin plots) ───────────────────────
print("=" * 65)
print("  17. Per-Emotion Latent Distribution (μ and σ)")
print("=" * 65)

N_DIST = 200

emo_zmeans = {}
emo_zstds  = {}

for idx, emo in enumerate(EMOTION_LABELS):
    emo_paths_d = test_paths[test_labels == idx][:N_DIST]
    imgs        = np.array([load_image(p).numpy() for p in emo_paths_d])
    z_m, z_lv, _ = encoder.predict(imgs, verbose=0)
    emo_zmeans[emo] = z_m
    emo_zstds[emo]  = np.exp(0.5 * z_lv)

summary_mu  = {e: np.mean(np.abs(emo_zmeans[e]), axis=1) for e in EMOTION_LABELS}
summary_std = {e: np.mean(emo_zstds[e],          axis=1) for e in EMOTION_LABELS}

fig, axes = plt.subplots(1, 2, figsize=(18, 6))
fig.suptitle("Per-Emotion Latent Distribution Summary",
             fontsize=14, fontweight='bold')

for ax, data, title, ylab in zip(
        axes,
        [summary_mu, summary_std],
        ['Mean |μ| across Latent Dims', 'Mean σ across Latent Dims'],
        ['|Mean Activation|', 'Standard Deviation']):

    plot_data = [data[e] for e in EMOTION_LABELS]
    positions = range(1, len(EMOTION_LABELS) + 1)

    vp = ax.violinplot(plot_data, positions=positions,
                       showmeans=True, showmedians=False, widths=0.7)

    for pc, col in zip(vp['bodies'], [EMOTION_COLORS[e] for e in EMOTION_LABELS]):
        pc.set_facecolor(col)
        pc.set_alpha(0.7)
        pc.set_edgecolor('white')
    for part in ('cbars', 'cmins', 'cmaxes', 'cmeans'):
        if part in vp:
            vp[part].set_color('#333')
            vp[part].set_linewidth(1.5)

    ax.set_xticks(positions)
    ax.set_xticklabels([e.capitalize() for e in EMOTION_LABELS],
                       rotation=30, ha='right')
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_ylabel(ylab)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('latent_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── 21b. Visual summary card ─────────────────────────────────────────────────
fig = plt.figure(figsize=(22, 14), facecolor='#0D1117')
gs  = gridspec.GridSpec(3, 4, figure=fig, hspace=0.55, wspace=0.4,
                        left=0.06, right=0.97, top=0.88, bottom=0.08)

# Title banner
ax_title = fig.add_subplot(gs[0, :])
ax_title.set_facecolor('#0D1117')
ax_title.axis('off')
ax_title.text(0.5, 0.80,
              "Emotion VAE  ·  FER2013  ·  Project Summary",
              ha='center', va='center', fontsize=24, fontweight='bold',
              color='white', transform=ax_title.transAxes)
ax_title.text(0.5, 0.35,
              f"Latent Dim: {LATENT_DIM}   |   Epochs: {EPOCHS}   |   "
              f"Final Recon Loss: {history.history['recon_loss'][-1]:.2f}   |   "
              f"k-NN Accuracy: {overall_acc * 100:.1f}%   |   "
              f"MSE + SSIM + β-KL Loss",
              ha='center', va='center', fontsize=12, color='#8B9BB4',
              transform=ax_title.transAxes)


def style_ax(ax, title):
    ax.set_facecolor('#161B22')
    ax.set_title(title, color='white', fontsize=10, fontweight='bold', pad=6)
    ax.tick_params(colors='#8B9BB4', labelsize=7)
    for sp in ax.spines.values():
        sp.set_color('#30363D')


# Reconstruction loss
ax1 = fig.add_subplot(gs[1, 0])
style_ax(ax1, 'Reconstruction Loss')
ax1.plot(history.history['recon_loss'], color='#F4A261', linewidth=2)
ax1.set_xlabel('Epoch', color='#8B9BB4', fontsize=8)
ax1.fill_between(range(len(history.history['recon_loss'])),
                 history.history['recon_loss'], alpha=0.15, color='#F4A261')

# KL loss
ax2 = fig.add_subplot(gs[1, 1])
style_ax(ax2, 'KL Loss  (with warm-up)')
ax2.plot(history.history['kl_loss'], color='#2A9D8F', linewidth=2)
ax2.axvline(x=WARMUP_EPOCHS - 1, color='#E63946', linestyle='--',
            linewidth=1.2, label=f'Warm-up end (ep {WARMUP_EPOCHS})')
ax2.legend(fontsize=7, labelcolor='#8B9BB4', facecolor='#161B22',
           edgecolor='#30363D')
ax2.set_xlabel('Epoch', color='#8B9BB4', fontsize=8)
ax2.fill_between(range(len(history.history['kl_loss'])),
                 history.history['kl_loss'], alpha=0.15, color='#2A9D8F')

# Per-emotion MSE
ax3 = fig.add_subplot(gs[1, 2])
style_ax(ax3, 'Per-Emotion Recon MSE (Test)')
means_plot = [avg_mse_per_emo[e][0] for e in EMOTION_LABELS]
stds_plot  = [avg_mse_per_emo[e][1] for e in EMOTION_LABELS]
ax3.bar(range(len(EMOTION_LABELS)), means_plot,
        color=[EMOTION_COLORS[e] for e in EMOTION_LABELS],
        yerr=stds_plot, capsize=3,
        error_kw={'ecolor': '#aaa', 'elinewidth': 1},
        edgecolor='none')
ax3.set_xticks(range(len(EMOTION_LABELS)))
ax3.set_xticklabels([e[:3].capitalize() for e in EMOTION_LABELS],
                    color='#8B9BB4', fontsize=7)

# k-NN per-class accuracy
ax4 = fig.add_subplot(gs[1, 3])
style_ax(ax4, f'k-NN Accuracy per Class  (k=5)')
per_class_acc = [cm_norm[i, i] * 100 for i in range(len(EMOTION_LABELS))]
bars_acc = ax4.barh(range(len(EMOTION_LABELS)), per_class_acc,
                    color=[EMOTION_COLORS[e] for e in EMOTION_LABELS],
                    edgecolor='none')
ax4.set_yticks(range(len(EMOTION_LABELS)))
ax4.set_yticklabels([e.capitalize() for e in EMOTION_LABELS],
                    color='#8B9BB4', fontsize=8)
ax4.set_xlim(0, 110)
ax4.set_xlabel('Accuracy (%)', color='#8B9BB4', fontsize=8)
for bar, val in zip(bars_acc, per_class_acc):
    ax4.text(val + 1, bar.get_y() + bar.get_height() / 2,
             f'{val:.1f}%', va='center', color='#8B9BB4', fontsize=7)

# Metrics table
ax_table = fig.add_subplot(gs[2, :])
ax_table.set_facecolor('#0D1117')
ax_table.axis('off')

table_data = [
    ["Emotion",   "Avg MSE (×10⁻³)", "Avg SSIM", "k-NN Acc (%)",
     "Prototype Cos-Sim\n(to nearest)", "Latent Dist\n(Euclidean, to nearest)"],
]

# Find nearest neighbor for each emotion (excluding itself)
cos_sim_proto = 1 - pairwise_distances(
    np.stack([emotion_prototypes[e] for e in EMOTION_LABELS]), metric='cosine')
eucl_proto    = pairwise_distances(
    np.stack([emotion_prototypes[e] for e in EMOTION_LABELS]), metric='euclidean')

for i, emo in enumerate(EMOTION_LABELS):
    sim_row   = cos_sim_proto[i].copy(); sim_row[i] = -999
    dist_row  = eucl_proto[i].copy();   dist_row[i] = 999
    near_sim  = EMOTION_LABELS[np.argmax(sim_row)]
    near_dist = EMOTION_LABELS[np.argmin(dist_row)]
    table_data.append([
        emo.capitalize(),
        f"{avg_mse_per_emo[emo][0] * 1000:.2f}",
        f"{avg_ssim_per_emo[emo][0]:.4f}",
        f"{cm_norm[i, i] * 100:.1f}%",
        f"{np.max(sim_row):.3f} ({near_sim})",
        f"{np.min(dist_row):.1f} ({near_dist})",
    ])

tbl = ax_table.table(
    cellText=table_data[1:],
    colLabels=table_data[0],
    cellLoc='center',
    loc='center',
    bbox=[0.01, 0.02, 0.98, 0.94],
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9.5)

for (r, c), cell in tbl.get_celld().items():
    cell.set_edgecolor('#30363D')
    cell.set_linewidth(0.8)
    if r == 0:
        cell.set_facecolor('#21262D')
        cell.set_text_props(color='white', fontweight='bold')
    else:
        emo = EMOTION_LABELS[r - 1]
        cell.set_facecolor('#161B22' if r % 2 == 0 else '#0D1117')
        cell.set_text_props(
            color=EMOTION_COLORS[emo] if c == 0 else '#C9D1D9'
        )

fig.suptitle("", fontsize=1)   # suppress default suptitle space
plt.savefig('project_summary_card.png', dpi=150, bbox_inches='tight',
            facecolor='#0D1117')
plt.show()

print("\n  ✓  project_summary_card.png saved")
print("\n" + "═" * 65)
print("  All visualizations complete!")
print("  Files saved:")
saved_files = [
    'sample_images.png', 'training_curves.png',
    'reconstructions.png', 'per_class_mse.png',
    'latent_space.png', 'prototype_heatmap.png',
    'vector_arithmetic.png',
    'walk_angry_happy.png', 'walk_sad_surprise.png',
    'walk_disgust_neutral.png', 'walk_fear_happy.png',
    'analogies.png',
    'recon_error_by_emotion.png', 'latent_confusion_matrix.png',
    'latent_distributions.png', 'best_worst_reconstructions.png',
    'inter_emotion_distances.png', 'kl_recon_tradeoff.png',
    'project_summary_card.png',
]
for f in saved_files:
    print(f"    • {f}")
print("=" * 65)
```
